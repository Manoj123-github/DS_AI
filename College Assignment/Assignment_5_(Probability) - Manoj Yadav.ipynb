Q.2. In the game of snakes and ladders, a fair die is thrown. If event E1 represents all the events of getting a natural number less than 4, event E2 consists of all the events of getting an even number and E3 denotes all the events of getting an odd number. List the sets representing the following

1. E1 or E2 or E3
2. E1 and E2 and E3
3. E1 but not E3


```python
dice=[1,2,3,4,5,6]
E1=[]
E2=[]
E3=[]
for r in dice:
  r=int(r)
  if r<4:
    E1.append(r)
  if r%2==0:
    E2.append(r)
  else:
    E3.append(r)

def snakesladers(l,s,k):
  E1orE2orE3=[]
  E1andE2andE3=[]
  E1notE3=[]
  for i in l:
    if((i in E1orE2orE3)==False):
      E1orE2orE3.append(i)
    if((i not in s)==True and (i not in k)==True):
      E1andE2andE3.append(i)
  for i in s:
    if((i not in E1orE2orE3)==True):
      E1orE2orE3.append(i)
    if(((i in l)==True and (i in k)==True) and (i not in E1andE2andE3)==True):
      E1andE2andE3.append(i)
    if((i in l)==True and (i not in k)==True):
      E1notE3.append(i)
  for i in k:
    if((i not in E1orE2orE3)==True):
      E1orE2orE3.append(i)
    if(((i in l)==True and (i in s)==True) and (i not in E1andE2andE3)==True):
      E1andE2andE3.append(i)
  E1orE2orE3.sort()
  E1andE2andE3.sort()
  E1notE3.sort()
  print("E1 or E2 or E3: ",E1orE2orE3)
  print("E1 and E2 and E3: ",E1andE2andE3)
  print("E1 but not E3: ",E1notE3)

snakesladers(E1,E2,E3)
```

Q. 3. How many permutations of the letters of the word 'ARTICLE'  have consonants in the first and last positions?

```python
Word=["A","R","T","I","C","L","E"]

def consonants(word):  
  consonant=[]
  for r in word:
    if (r!="A" and r!="E" and r!="I" and r!="O" and r!="U"):
      consonant.append(r)
  return consonant

def fact(g):
  fact=1
  for s in range(1,g+1):
    fact=fact*s
  return fact

def permutations(Word):
  k=len(consonants(Word))
  n=len(Word)-2
  r=n
  permutation= fact(n)/fact(n-r)
  PmtWithConst=k*permutation*(k-1)
  print("Number of Permutaion=",PmtWithConst)

permutations(Word)
```

Q. 6. You have 4 math books and 5 history books to put on a shelf that has 5 slots. In how many ways can the books be shelved if the first three slots are filled with math books and the next two slots are filled with history books?

```python
def fact(g):
  fact=1
  for s in range(1,g+1):
    fact=fact*s
  return fact

def permutations(n,r):
  permutation= fact(n)/fact(n-r)
  return permutation

# As 4 math books must be filled in First three slots therefore it can be filled in 4P3 ways
r=permutations(4,3)
# As 5 History books must be filled in next two slots therefore it can be filled in 5P2 ways
n=permutations(5,2)
print("Number of ways the bookshelf can be filled=",r*n)
```

Q.8.  A Birthday Problem (Collision of Birth Days): 

"If there are ‘n’ people in a room, what is the chance that some pair among them have the same birthday?"

Sampling with and Without Replacement


```python
def prodprob(n,r):
  prob=1
  for g in range(1,r):
    prob= prob*((n-g)/n)
  return prob
#There are 365 days in year and ignore leap years as well as multiple births (twins, for example) 
D=365
#let persons in room be 30 (n=30)
P=5
#diffrent birthdays of persons in room
NDP=prodprob(D,P)
#same birthdays persons=1-diffrent birthdays of persons
SP=1-NDP
print("for 30 persons in room same birthday persons=",SP)
print("for 30 persons in room different birthday persons=",NDP)
```
