```python
# From csv
import pandas as pd
EmpId = [101,102,103,104,105,106,107,108,109,110]
FName = ['Manoj','Rahul','Satish','Anita','Pooja','Anish','ritvik','vikrant','megha','Darshana']
LName = ['Yadav','Rai','Dubey','Shindey','Jha','Tivari','thirki','kanchan','Yadav','Kanchan']
Age = [23,30,25,21,35,30,20,42,52,21]
Designation=['ProjectEngineer','Tester','Software','Webdeveloper','Manager','ProjectEngineer','Tester','Software','Webdeveloper','Manager']
BasicSalary = [40000,35000,50000,52000,60000,40700,35000,50250,52000,61000]
joblist=zip(EmpId,FName,LName,Age,Designation,BasicSalary)
df=pd.DataFrame(data=joblist,columns=['EmpId','FName','LName','Age','Designation','BasicSalary'])
df.to_csv('Job.csv',index=True,header=False)
df
```

Descriptive Statistics

```python
#mean
print('Mean of numeric columns-\n',df.mean())
```

Measures of Variability 

1. Standard Deviation

2. Variance

3. Min and Max Variables

4. Skewness and Kurtosis

5. Range

6. Percentiles/Quartiles

7. Box Plots


```python
#Standard deviation
print('standard deviation of numeric columns - \n',df.std())
```

```python
#variance
print('variance of numeric columns - \n',df.var())
```

```python
#Min & Max 
print('\nMin variables of all columns -\n',df.min())
print('\nMax variables of all columns -\n ',df.max())
```

```python
#Skewness
print('Skewness of numeric columns - \n',df.skew())
```

```python
#Kutosis
print('kurtosis of numeric columns - \n',df.kurtosis())
```

```python
#Range
print('Range of the column EmpId = ',df['EmpId'].max()-df['EmpId'].min())
print('Range of the column Age = ',df['Age'].max()-df['Age'].min())
print('Range of the column BasicSalary = ',df['BasicSalary'].max()-df['BasicSalary'].min())
```

```python
# Percentile
print("\nPercentile summary-")
print('\n\n10th Percentile of the column EmpId = ',np.percentile(df['EmpId'],10))
print('50th Percentile(medain) of the EmpId = ',np.percentile(df['EmpId'],50))
print('90th Percentile of the column EmpId = ',np.percentile(df['EmpId'],90))
print('\n\n10th Percentile of the column Age = ',np.percentile(df['Age'],10))
print('50th Percentile(medain) of the column Age = ',np.percentile(df['Age'],50))
print('90th Percentile of the column Age = ',np.percentile(df['Age'],90))
print('\n\n10th Percentile of the column BasicSalary = ',np.percentile(df['BasicSalary'],10))
print('50th Percentile(medain) of the column BasicSalary = ',np.percentile(df['BasicSalary'],50))
print('90th Percentile of the column BasicSalary = ',np.percentile(df['BasicSalary'],90))
```

```python
#qurtile
print("\nQuertile summary-")
print("\n25% quertile (Lower quertile) of numeric columns-\n ",df.quantile(q=0.25))
print("\n50% quertile (Middle quertile)of numeric columns-\n",df.quantile(q=0.5))
print("\n75% quertile (Upper quertile) of numeric columns-\n",df.quantile(q=0.75))
```
