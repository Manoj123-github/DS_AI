```
import pandas as pd
from sqlalchemy import create_engine
engine = create_engine('sqlite://', echo=False)
EmpId = ['E101','E102','E103','E104']
FName = ['Sweta','Manisha','Madhu','Anita']
LName = ['Subnis','Perdesi','Mali','Rai']
Designation = ['ProjectEngineer','Tester','SrSoftwareEngineer','WebDeveloper']
Basic_Salary = [40000,50000,60000,70000,80000]
Degrees = zip(EmpId,FName,FName,LName,Designation,Basic_Salary)
columns = ['EmpId','FName','FName','LName','Designation','Basic_Salary']
df = pd.DataFrame(data = Degrees,columns=columns)
df.to_sql('empdata', con=engine)
df

```
