```python
# Keywords Recommender 
##################################################

# the quality of recommender would be increased with better metadata
# we now build a recommender based on the following metadata: 
# the 3 top actors, the director, related genres and the movie plot keywords

# from our new features, cast, crew and keywords, we need to extract the three 
# most important actors, the director and the keywords associated with that movie
# our data is present in the form of "stringified" lists; 
# we need to convert them into a form that is usable to us
result = md.dtypes
print(result)
```

```python
# parse the stringified features into their corresponding python objects
features = ['cast', 'crew', 'keywords', 'genres']
from ast import literal_eval
for feature in features:
    md[feature] = md[feature].apply(literal_eval)
    
result = md.dtypes 
print(result)
```

```python
# get the director's name from the crew feature
# if director is not listed, return NaN
def get_director(x):
    for i in x:
        if i['job'] == 'Director':
            return i['name']
    return np.nan
```

```python
# define new director in a suitable form
md['director'] = md['crew'].apply(get_director)
```

```python
# returns the list top 5 elements or entire list; whichever is more.
def get_list(x):
    if isinstance(x, list):
        names = [i['name'] for i in x]
        #Check if more than 5 elements exist. If yes, return only first three. If no, return entire list.
        if len(names) > 5:
            names = names[:5]
        return names
    # return empty list in case of missing/malformed data
    return []
```

```python
# define cast, genres and keywords features that are in a suitable form
features = ['cast', 'keywords', 'genres']
md['str_cast'] = md['cast'].apply(get_list)
md['str_keys'] = md['keywords'].apply(get_list)
md['str_genr'] = md['genres'].apply(get_list)
```

```python
# print the new features
print(md[['title', 'director', 'str_cast', 'str_keys', 'str_genr']].head())
```

```python
# the next step would be to convert the names and keyword instances into lowercase 
# and strip all the spaces between them; this is done so that your vectorizer 
# doesn't count "Johnny Depp" and "Johnny Galecki" as the same. After this step,
# keywrods will be represented as "johnnydepp" and "johnnygalecki" and will be distinct 

# function to convert all strings to lower case and strip names of spaces
def cleanData(x):
    if isinstance(x, list):
        return [str.lower(i.replace(" ", "")) for i in x]
    else:
        #Check if director exists. If not, return empty string
        if isinstance(x, str):
            return str.lower(x.replace(" ", ""))
        else:
            return ''
```

```python
# apply clean_data function to your features.
features = ['director', 'str_cast', 'str_keys', 'str_genr']
for feature in features:
    md[feature] = md[feature].apply(cleanData)
```

```python
# print the new features
print(md[['title', 'director', 'str_cast', 'str_keys', 'str_genr']].head())
```

```python
# we now create our "metadata soup", which is a string that contains all the 
# metadata that is to be given to your vectorizer (namely actors, director and keywords)

# function
def createSoup(x):
    return ' '.join(x['str_keys']) + ' ' + ' '.join(x['str_cast']) + ' ' + x['director'] + ' ' + ' '.join(x['str_genr'])

```

```python
# create a new soup feature
md['soup'] = md.apply(createSoup, axis=1)
print(md['soup'][0])
print(md['soup'].head())
```

```python
# the next steps are the same as what you did with your content based recommender
# important difference is that we use the CountVectorizer() instead of TF-IDF

# import CountVectorizer and create the count matrix
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# use CountVectorizer
count = CountVectorizer(stop_words='english')
count_matrix = count.fit_transform(md['soup'])
print(count_matrix.shape)
print(count_matrix[5])
```

```python
 #compute the Cosine Similarity matrix based on the count_matrix
cosSimilarity = cosine_similarity(count_matrix, count_matrix)
```

```python
# reset index of your main DataFrame and construct reverse mapping as before
md = md.reset_index()
indices = pd.Series(md.index, index=md['title'])
```

```python
# function that takes in movie title as input and outputs most similar movies
def getReccos(title, cosine_sim=cosSimilarity):
    # Get the index of the movie that matches the title
    try:
        idx = indices[title]
    except:
        return "Title Not Found ..."
    # Get the pairwsie similarity scores of all movies with that movie
    sim_scores = list(enumerate(cosine_sim[idx]))
    # Sort the movies based on the similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    # Get the scores of the 10 most similar movies
    sim_scores = sim_scores[1:11]
    # Get the movie indices
    movie_indices = [i[0] for i in sim_scores]
    # Return the top 10 most similar movies
    return md['title'].iloc[movie_indices],sim_scores
```
