```python
# imports
import requests
from requests import get
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import re
import time
from random import randint
```

```python
# url header
headers = {"Accept-Language": "en-US, en;q=0.5"}
url = "https://www.imdb.com/search/title/?groups=top_1000&ref_=adv_prv"
#results = requests.get(url, headers=headers)
```

```python
# first url
# https://www.imdb.com/search/title/?groups=top_1000&ref_=adv_prv
# next url
# https://www.imdb.com/search/title/?title_type=feature,tv_series&count=100&start=101&ref_=adv_nxt
# after that the url
# https://www.imdb.com/search/title/?title_type=feature,tv_series&count=100&start=201&ref_=adv_nxt
```

```python
# init lists
movieTitle = []
movieDate = []
movieRuntime = []
movieGenre = []
movieRating = []
movieScore = []
movieDescription = []
movieDirector = []
movieStars = []
movieVotes = []
movieGross = []
```

```python
# page range
pages = np.arange(1, 201, 50)
```

```python
# loop through all pages
for page in pages: 
    
    # message
    print("\n==============================")
    print("Loading Movies "+str(page)+" to "+str(page+49))
    print("==============================")

    # get url
    results = requests.get("https://www.imdb.com/search/title/?groups=top_1000&start=" + str(page) + "&ref_=adv_nxt", headers=headers)

    # soup object
    soup = BeautifulSoup(results.text, "lxml")

    # sleep    
    time.sleep(randint(2,10))
    
    # top level movie frame
    movieFrame = soup.find_all('div', class_='lister-item mode-advanced')
    
    # loop through all movies in the movie frame
    for movie in movieFrame:

        # movie start
        print("==============================")
            
        # first line
        movieFirstLine = movie.find("h3", class_="lister-item-header")
        
        # title
        try:
            mTitle = movieFirstLine.find("a").text
        except:
            mTitle = None
        print(mTitle)
        movieTitle.append(mTitle)
    
        # date / year
        try:
            mDate = movieFirstLine.find_all("span")[-1].text
            #mDate = mdate.replace("\(","").split(",")
            #mDate = mdate.replace("\)","").split(",")
            mDate = re.sub(r"[()]","", mDate)
        except:
            mDate = None
        print(mDate)        
        movieDate.append(mDate)
        
        # runtime    
        try:
            mRuntime = movie.find("span", class_="runtime").text[:-4]
        except:
            mRuntime = None 
        print(mRuntime)        
        movieRuntime.append(mRuntime)
    	
        # genre
        try:
            mGenre = movie.find("span", class_="genre")
            mGenre = mGenre.text.rstrip()
            mGenre = mGenre.replace("\n","")
            mGenre = mGenre.split(",")
        except:
            mGenre = None 
        print(mGenre)        
        movieGenre.append(mGenre)
    
        # rating	
        try:
            mRating = movie.find("strong").text
        except:
            mRating = None
        print(mRating)        
        movieRating.append(mRating)
    
        # score	
        try:
            mScore = movie.find("span", class_="metascore favorable")
            mScore = mScore.text.strip()
        except:
            mScore = None
        print(mScore)        
        movieScore.append(mScore)
    	
        # description
        try:
            mDescription = movie.find_all("p", class_="text-muted")[-1]
            mDescription = mDescription.text.strip()
        except:
            mDescription = None
        print(mDescription)        
        movieDescription.append(mDescription)
    
        # cast
        try:
            mCasts = movie.find("p", class_="")
            mCasts = mCasts.text.replace("\n","")
            mCasts = mCasts.split('|')
            mCasts = [x.strip() for x in mCasts]
            mCasts[0] = mCasts[0].replace("Director:","")
            mCasts[1] = mCasts[1].replace("Stars:","")
            mDirector = mCasts[0]
            mStars = mCasts[1:]
            mStars = ",".join(mStars)
        except:
            mCast = None        
            mDirector = None
            mStars = None
        print(mCasts)
        print(type(mCasts))
        print(mDirector)      
        print(mStars)
        print(type(mStars))
        movieDirector.append(mDirector)
        movieStars.append(mStars)
        
        # there are two NV containers, grab both of them as they hold both the votes and the grosses
        try:
            mNV = movie.find_all("span", attrs={"name": "nv"})
            if len(mNV) == 2:
                mVotes = mNV[0].text
                mGross = mNV[1].text
            elif len(mNV) == 1:
                mVotes = mNV[0].text
                mGross = None
            else:
                mVotes = None
                mGross = None
        except:
            mVotes = None
            mGross = None
        print(mVotes)
        print(mGross)
        movieVotes.append(mVotes)
        movieGross.append(mGross)
```

```python
# pandas dataframe        
movies = pd.DataFrame({
'MovieTitle': movieTitle,
'MovieDate': movieDate,
'MovieRuntime': movieRuntime,
'MovieRating': movieRating,
'MovieScore': movieScore,
'MovieDescription': movieDescription,
'MovieDirector': movieDirector,
'MovieStars': movieStars,
'MovieVotes': movieVotes,
'MovieGross': movieGross,
})
```
