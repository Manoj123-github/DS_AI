```python
# C is the mean vote across the whole report
C = md['vote_average'].mean()
print(C)
```

```python
# m is the minimum votes required to be listed in the chart
m = md['vote_count'].quantile(0.50)
print(m)

# create a copy and filter for m
# Filter out all qualified movies into a new DataFrame
tm = md.copy()
tm = tm[tm['vote_count'] >= m]
print(type(tm))
print(tm.shape)
print(tm.head())
```

```python
# function that computes the weighted rating of each movie
def weightedRating(x, m=m, C=C):
    v = x['vote_count']
    R = x['vote_average']
    # Calculation based on the IMDB formula
    return (v/(v+m) * R) + (m/(m+v) * C)
```

```python
# define a new feature 'score' and calculate its value with `weighted_rating()`
tm['score'] = tm.apply(weightedRating, axis=1)
print(type(tm))
print(tm.shape)
print(tm.head())
```

```python
# sort movies based on score calculated above
tm = tm.sort_values('score', ascending=False)
```

```python
# print the top 15 movies
print(tm[['title', 'vote_count', 'vote_average', 'score']].head(15))
```

```python
# handle genres wise top n
print(tm['genres'].head())
# this is dictionary ... convert to list
tm['genres'] = tm['genres'].fillna('[]').apply(eval).apply(lambda x: [i['name'] for i in x] if isinstance(x, list) else [])
print(tm['genres'].head())
```

```python
lsGenres = ['Action','Adventure','Animation","Children','Comedy', 'Crime','Documentary','Drama','Fantasy','Film-Noir','Horror','Musical','Mystery','Romance','Sci-Fi','Thriller','War','Western']
print(lsGenres)
```

```python
# movies of a genre
def getGenreMovies(pGenre):
    global tm
    mask = tm['genres'].apply(lambda x: pGenre in x)
    #print(mask[0:5])
    mg = tm[mask]
    # print the top 5 movies
    print("\nGenre:", pGenre)
    print("==============================================")
    print(mg[['title', 'vote_count', 'vote_average', 'score']].head(5))
```

```python
# sample genre
getGenreMovies('Comedy')
```

```python
# get top 5 for all genres
for stGenre in lsGenres:
    getGenreMovies(stGenre)
```
